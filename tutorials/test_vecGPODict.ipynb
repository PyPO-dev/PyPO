```python
import numpy as np

from PyPO.Checks import check_vecGPODict, calc_beam
from PyPO.Enums import Units
```

```python
# Set up a Gaussian beam with known parameters

lam = 1.0 * Units.MM

k = 2*np.pi/lam

w0 = 5.0*lam
z = 200.0*lam

zc = k*w0**2

w_ref = w0*np.sqrt(1 + ((lam*z)/(np.pi*w0**2))**2)
R_ref = z*(1 + ((np.pi*w0**2)/(lam*z))**2)
```

```python
print(f"Input lambda : {lam:.6f} mm")
print(f"Input w0     : {w0:.6f} mm")
print(f"Input z      : {z:.6f} mm")
print(f"Reference w  : {w_ref:.6f} mm")
print(f"Reference R  : {R_ref:.6f} mm")
print(f"Z_c          : {zc:.6f} mm")
```

```python
vecGPObeam_w0z = {
        "name"      : "testGaussBeam_w0z",
        "lam"       : lam,
        "w0"        : w0,
        "z"         : z,
        "n"         : 1,
        "power"     : 1
}

vecGPObeam_w0w = {
        "name"      : "testGaussBeam_w0w",
        "lam"       : lam,
        "w0"        : w0,
        "w"         : w_ref,
        "n"         : 1,
        "power"     : 1
}

vecGPObeam_w0R = {
        "name"      : "testGaussBeam_w0R",
        "lam"       : lam,
        "w0"        : w0,
        "R"         : R_ref,
        "n"         : 1,
        "power"     : 1
}

vecGPObeam_zw = {
        "name"      : "testGaussBeam_zw",
        "lam"       : lam,
        "z"         : z,
        "w"         : w_ref,
        "n"         : 1,
        "power"     : 1
}

vecGPObeam_zR = {
        "name"      : "testGaussBeam_zR",
        "lam"       : lam,
        "z"         : z,
        "R"         : R_ref,
        "n"         : 1,
        "power"     : 1
}

vecGPObeam_wR = {
        "name"      : "testGaussBeam_wR",
        "lam"       : 1.0,
        "w"         : w_ref,
        "R"         : R_ref
}

indicts = [
    vecGPObeam_w0z,
    vecGPObeam_w0w,
    vecGPObeam_w0R,
    vecGPObeam_zw,
    vecGPObeam_zR,
    vecGPObeam_wR
]
```
