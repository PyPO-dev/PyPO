# Vectorial Gaussian Beams

Per Cullen & Yu "Complex Source Point Theory of the Electromagnetic Open Resonator" (1979), we use the complex source point representation of a Gaussian beam to generate the E- and H- fields of a Gaussian beam.

The complex source point method generates a Gaussian beam travelling along the positive z-axis by displacing a combined electric and magnetic dipole (oriented along the x- and y-axes respectively) by an imaginary distance $i z_c$ in the positive z direction.

$z_c$ is the confocal distance of the beam, related to the wavelength $\lambda$ and the beamwaist radius $w_0$ by:

$$ z_c = \frac{\pi w_0^2}{\lambda}$$


## Symmetrization

The magnetic dipole oriented along the y-axis is introduced in order to provide symmetry between the E- and H-fields, so that the fields satisfy the duality principle, and are invariant under the substitions

$$ \mathbf{E} \rightarrow -Z \mathbf{H} $$

$$ \mathbf{H} \rightarrow \frac{1}{Z} \mathbf{E} $$

$$ \mathbf{J} \rightarrow -\frac{1}{Z} \mathbf{M} $$

$$ \mathbf{M} \rightarrow Z \mathbf{J} $$


The strength $p_m$ of the magnetic dipole required to symmetrize an electric dipole of strengh $p_e$ is given by

$$ p_m = - p_e \sqrt{\epsilon / \mu} = - p_e / Z $$

where $Z$ is the impedance of freespace.


# Symmetrized vector Gaussian Beam

Assuming a time factor of $e^{i \omega t}$, according to Cullen & Yu, the fields due to the electric and magnetic dipoles at the origin are:

$$ E_x = \frac{k^2}{r} e^{-ikr}\Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{k^2}{r} e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ E_z = \frac{k^2}{r} e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = - \frac{k^2}{Z r} e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{k^2}{Z r} e^{-ikr}\Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = \frac{k^2}{Zr} e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{y}{r}\Bigg] $$

Comparison of these equations with those for the GRASP fields due to a current at a point, where a time factor of $e^{-i \omega t}$ is assumed:

$$ \mathbf{E} = \frac{k^2}{4\pi} \Bigg\{ Z \Bigg[J \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{J}\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

$$ - \mathrm{M}\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{-ikR} $$

and 

$$ \mathbf{H} = \frac{k^2}{4\pi} \Bigg\{ \frac{1}{Z} \Bigg[M \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{M}\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

$$ + \mathrm{J}\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{-ikR} $$

as well as the time reversed versions ($e^{i \omega t}$):

$$ \mathbf{E} = \frac{k^2}{4\pi} \Bigg\{ Z \Bigg[J \Bigg( \frac{-i}{kR} + \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{J}\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} - \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

$$ + \mathrm{M}\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{ikR} $$

and 

$$ \mathbf{H} = \frac{k^2}{4\pi} \Bigg\{ \frac{1}{Z} \Bigg[M \Bigg( \frac{-i}{kR} + \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{M}\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} - \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

$$ - \mathrm{J}\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{ikR} $$

shows that the Cullen & Yu equations correspond to our time reversed case.

Our forward time direction fields for a symmetrized Gaussian beam are therefore:

$$ E_x = \frac{k^2}{4 \pi} e^{ikr}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3i}{(kr)^2} - \frac{3}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{k^2}{4 \pi} e^{ikr}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ E_z = \frac{k^2}{4 \pi} e^{ikr}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3i}{(kr)^2} - \frac{3}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = - \frac{k^2}{4 \pi Z} e^{ikr}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{k^2}{4 \pi Z} e^{ikr}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3i}{(kr)^2} - \frac{3}{(kr)^3}\bigg)\frac{y^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = \frac{k^2}{4 \pi Z} e^{ikr}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3i}{(kr)^2} - \frac{3}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{y}{r}\Bigg] $$

# Normalization

The above equations give results with very large or small magnitudes for the fields due to the imaginary part of $r$ giving the exponential of a real number in $e^{±i k r}$.  Per Cullen & Yu, we therefore multiply the fields by a factor $ e^{-k z_c} $, so that:

$$ e^{±i k r} \rightarrow e^{\pm i k r \mp k z_c} $$

# Implementation

We now implement each of these fields in numpy

## Define the Gaussian Beam

```python
# Setting up parameters for defining simulation.
# Source parameters and distances
lam = 1*Units.MM                # Wavelength of light in mm
k = 2*np.pi/lam
print(f"Wavenumber                : {k:7.3f} mm")

w0 = 5*lam  # Gaussian beamwaist

r_source = 2.5*w0               # Radius of disk in mm
print(f"Radius of source grid     : {r_source:7.3f} mm")

zc = np.pi*w0**2 / lam # Confocal distance in mm
print(f"Confocal distance         : {zc:7.3g} mm")

E0 = np.sqrt(8/(w0**2)) # Normalise to 4pi Watts
print(f"Field normalization       : {E0:7.3g} √Watt")

theta_0 = np.rad2deg(np.atan(lam/(np.pi*w0))) # Calculate far-field divergence angle
print(f"Farfield divergence angle : {theta_0:7.3f}°")

fwhm = 1.18*theta_0
print(f"Farfield FWHM angle       : {fwhm:7.3f}°")

# Calculate the size at the test point
z_grid = 0
print(f"Near-field grid distance     : {z_grid:7.3f} mm")

w_grid = w0*np.sqrt(1 + ((lam*z_grid)/(np.pi*w0**2))**2)
print(f"Beamwidth at near-field grid : {w_grid:7.3f} mm")
```

```python
import numpy as np
import matplotlib.pyplot as plt
    
from PyPO.System import System
from PyPO.Enums import Units
```

```python
# Create a system

s = System()
```

```python
# Create a source grid
source_plane = {
        "name"      : "source_plane",
        "gmode"     : "uv",
        # "lims_x"    : np.array([-r_disk, r_disk]),
        # "lims_y"    : np.array([-r_disk, r_disk]),
        "lims_u"    : np.array([0, r_source]),
        "lims_v"    : np.array([0, 360]),
        "gridsize"  : np.array([101, 360])
        }

s.addPlane(source_plane)

s.translateGrids("source_plane", np.array((0, 0, z_grid)))

source_grid = s.generateGrids("source_plane")
```

## Electric field

```python
# Precalculate r

r = np.sqrt(source_grid.x**2 + source_grid.y**2 + (source_grid.z + 1j*zc)**2)
```

```python
r
```

### $E_x$ component

$$ E_x = \frac{E_0 z_c}{2}\frac{k^2}{4 \pi} e^{ikr - kz_c}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3i}{(kr)^2} - \frac{3}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$



```python
prefactor = - E0 * k**2 / (4*np.pi) * np.exp(-1j * k * r - k*zc) * zc / 2
```

```python
term1 = - 1j/(k*r) - 1/(k*r)**2 + 1j/(k*r)**3
term2 = (1j/(k*r) + 3/(k*r)**2 - 3j/(k*r)**3) * (source_grid.x**2 / r**2)
term3 = (1j/(k*r) + 1 /(k*r)**2) * (source_grid.z + 1j*zc) / r
```

```python
Ex = prefactor * (term1 + term2 - term3)
```

```python
efield = Ex

grid = source_grid

vmax = 20*np.log10(np.max(np.abs(efield)))

print(vmax)

vmin = vmax - 40

x_grid = grid.x
y_grid = grid.y
z_grid = grid.z
phase_factor = np.exp(1j*2*np.pi/lam*z_grid)
#phase_factor = 1
fig, ax = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'wspace':0.5})

ampfig = ax[0].pcolormesh(x_grid, y_grid, 20*np.log10(np.abs(efield)) - vmax)
phasefig = ax[1].pcolormesh(x_grid, y_grid, np.angle(efield))

c1 = fig.colorbar(ampfig, orientation='vertical')
c2 = fig.colorbar(phasefig, orientation='vertical')

ax[0].set_aspect(1)
ax[1].set_aspect(1)
```

```python
row = 0

xlabel = 'Radius (mm)'

vrange = 40

vmax = 20*np.log10(np.max(np.abs(efield)))
print(vmax)
vmin = vmax - vrange

fig, ax = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'wspace':0.5})
ax[0].plot(grid.x[:,row], 20*np.log10(np.abs(efield[:,row])))
ax[0].axvline(w_grid, ls=":")
ax[0].axhline(vmax+20*np.log10(np.exp(-1)), ls=":")
ax[0].set_ylim((vmin, vmax+1))
ax[0].set_xlabel(xlabel)
ax[0].set_ylabel('Scalar field power (dBi)')

ax[1].plot(grid.x[:,row], np.unwrap(np.angle(efield[:,row])))
ax[1].set_xlabel(xlabel)
ax[1].set_ylabel('Scalar field Phase (rad)')
```

```python
### Check the normalization

np.sum(np.abs(Ex)**2 * np.deg2rad(1)) / (4*np.pi)
```

### $E_y$ component

$$ E_y = - \frac{E_0 z_c}{2} \frac{k^2}{4 \pi} e^{ikr}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^2}\bigg)\frac{xy}{r^2} $$

```python
Ey = - prefactor*( 1j / (k*r) + 3/ (k*r)**2 - 3j/(k*r)**3) * source_grid.x * source_grid.y / r**2
```

```python
efield = Ey

grid = source_grid

vmax = 20*np.log10(np.max(np.abs(Ex)))

print(vmax)

vmin = vmax - 80

x_grid = grid.x
y_grid = grid.y
z_grid = grid.z
phase_factor = np.exp(1j*2*np.pi/lam*z_grid)
#phase_factor = 1
fig, ax = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'wspace':0.5})

ampfig = ax[0].pcolormesh(x_grid, y_grid, 20*np.log10(np.abs(efield)) - vmax, vmax=vmax, vmin=vmin)
phasefig = ax[1].pcolormesh(x_grid, y_grid, np.angle(efield))

c1 = fig.colorbar(ampfig, orientation='vertical')
c2 = fig.colorbar(phasefig, orientation='vertical')

ax[0].set_aspect(1)
ax[1].set_aspect(1)
```

### $E_z$ component

$$ E_z = \frac{E_0 z_c}{2} \frac{k^2}{4 \pi} e^{ikr - kz_c}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

```python
term1 = (1j/(k*r) + 3/(k*r)**2 - 3j/(k*r)**3) * source_grid.x * (source_grid.z + 1j*zc) / r**2
```

```python
term2 = ( 1j/(k*r)**2 + 1/(k*r)**3) * source_grid.x / r
```

```python
Ez = prefactor*(term1 - term2)
```

```python
efield = Ez

grid = source_grid

vmax = 20*np.log10(np.max(np.abs(Ex)))

print(vmax)

vmin = vmax - 120

x_grid = grid.x
y_grid = grid.y
z_grid = grid.z
phase_factor = np.exp(1j*2*np.pi/lam*z_grid)
#phase_factor = 1
fig, ax = plt.subplots(1,2, figsize=(10,5), gridspec_kw={'wspace':0.5})

ampfig = ax[0].pcolormesh(x_grid, y_grid, 20*np.log10(np.abs(efield)) - vmax, vmax=vmax, vmin=vmin)
phasefig = ax[1].pcolormesh(x_grid, y_grid, np.angle(efield))

c1 = fig.colorbar(ampfig, orientation='vertical')
c2 = fig.colorbar(phasefig, orientation='vertical')

ax[0].set_aspect(1)
ax[1].set_aspect(1)
```
