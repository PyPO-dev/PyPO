# Vectorial Gaussian Beams

Per Cullen & Yu "Complex Source Point Theory of the Electromagnetic Open Resonator" (1979), we use the complex source point representation of a Gaussian beam to generate the E- and H- fields of a Gaussian beam.

The complex source point method generates a Gaussian beam travelling along the positive z-axis by displacing a combined electric and magnetic dipole (oriented along the x- and y-axes respectively) by an imaginary distance $i z_c$ in the positive z direction.

$z_c$ is the confocal distance of the beam, related to the wavelength $\lambda$ and the beamwaist radius $w_0$ by:

$$ z_c = \frac{\pi w_0^2}{\lambda}$$


## Symmetrization

The magnetic dipole oriented along the y-axis will be introduced in order to provide symmetry between the E- and H-fields, so that the fields satisfy the duality principle, and are invariant under the substitions

$$ \mathbf{E} \rightarrow -Z \mathbf{H} $$

$$ \mathbf{H} \rightarrow \frac{1}{Z} \mathbf{E} $$

$$ \mathbf{J} \rightarrow -\frac{1}{Z} \mathbf{M} $$

$$ \mathbf{M} \rightarrow Z \mathbf{J} $$


The strength $p_m$ of the magnetic dipole required to symmetrize an electric dipole of strengh $p_e$ is given by

$$ p_m = - p_e \sqrt{\epsilon / \mu} = - p_e / Z $$

where $Z$ is the impedance of freespace.


# Cullen & Yu Gaussian Beam

Assuming a time factor of $e^{i \omega t}$, according to Cullen & Yu, the fields due to the electric and magnetic dipoles at the origin are:

$$ E_x = \frac{k^2}{r} e^{-ikr}\Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{k^2}{r} e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ E_z = - \frac{k^2}{r} e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = - \frac{k^2}{Z r} e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{k^2}{Z r} e^{-ikr}\Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = - \frac{k^2}{Zr} e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{y}{r}\Bigg] $$

## Normalization

Cullen & Yu normalize these fields by multiplying by a factor

$$ -i \frac{z_c}{ 2 k^2} e^{-k z_c} $$

which gives a field strength of unity at $(0, 0, 0)$ for $E_x$.

yielding:

$$ E_x = - i \frac{z_c}{2 r} e^{-k z_c} e^{-ikr} \Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - (-i) \frac{z_c}{2 r} e^{-k z_c}  e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ E_z = - (-i) \frac{z_c}{2 r} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{x(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = i \frac{z_c}{2 Z r} e^{-k z_c}  e^{-ikr}\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{xy}{r^2} $$

$$ H_y = - i \frac{z_c}{2 Z r} e^{-k z_c} e^{-ikr}\Bigg[\bigg(1 - \frac{i}{kr} - \frac{1}{(kr)^2}\bigg) - \bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y^2}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = i \frac{z_c}{2 Z r} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(1 - \frac{3i}{kr} - \frac{3}{(kr)^2}\bigg)\frac{y(z + i z_c)}{r^2} + \bigg(1 - \frac{i}{kr}\bigg)\frac{y}{r}\Bigg] $$

Multiplying the inverse kr power sums by $-i/kr$ gives

$$ E_x = \frac{z_c k}{2} e^{-k z_c} e^{-ikr} \Bigg[\bigg(\frac{-i}{kr} - \frac{1}{(kr)^2} +  \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{z_c k}{2} e^{-k z_c}  e^{-ikr}\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ E_z = - \frac{z_c k}{2} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = - \frac{z_c k}{2 Z} e^{-k z_c}  e^{-ikr}\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{z_c k}{2 Z} e^{-k z_c} \frac{k^2}{Z r} e^{-ikr}\Bigg[\bigg(\frac{-i}{kr} - \frac{1}{(kr)^2} +  \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y^2}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = -\frac{z_c k }{2 Z} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y(z + i z_c)}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{y}{r}\Bigg] $$

We now substitute

$$ z_c = \frac{\pi w_0^2}{\lambda} = \frac{k w_0^2}{2} $$

$$ E_x = \frac{k^2 w_0^2}{4} e^{-k z_c} e^{-ikr} \Bigg[\bigg(\frac{-i}{kr} - \frac{1}{(kr)^2} +  \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{k^2 w_0^2}{4} e^{-k z_c}  e^{-ikr}\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ E_z = - \frac{k^2 w_0^2}{4} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = - \frac{k^2 w_0^2}{4 Z} e^{-k z_c}  e^{-ikr}\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{k^2 w_0^2}{4 Z} e^{-k z_c} e^{-ikr}\Bigg[\bigg(\frac{-i}{kr} - \frac{1}{(kr)^2} +  \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y^2}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = - \frac{k^2 w_0^2}{4 Z} e^{-k z_c}  e^{-ikr}\Bigg[\bigg(\frac{i}{kr} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y(z + i z_c)}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{y}{r}\Bigg] $$

# Derivation in PyPO field convention

Taking our fields for the source current dipole in our normalized fields convention, we move the source by a complex distance $i z_c$.

$$ \mathbf{E}_N = \frac{k^2}{4\pi} \Bigg\{ \mathbf{J}_N \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{J}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg) $$

$$ - \mathrm{M}_N\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{-ikR} $$


$$ \mathbf{H}_N = \frac{k^2}{4\pi} \Bigg\{\mathbf{M}_N \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + (\mathbf{M}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg) $$

$$ + \mathrm{J}_N\times\mathrm{\hat{R}}\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg\} e^{-ikR} $$




We now explicilty calculate the $\mathbf{J}_N$, $\mathbf{M}_N$, $(\mathbf{J}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}$, $(\mathbf{M}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}}$, $\mathbf{M}_N\times\mathbf{\hat{R}}$, and $\mathbf{J}_N\times\mathbf{\hat{R}}$ terms for a dipole of magnitude $J_x$ at a complex source point:

$$\mathbf{R} = x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}} $$



$$|\mathbf{R}| = R = \sqrt{x^2 + y^2 + (z + i z_c )^2} = (z + i z_c) \sqrt{1 + \frac{x^2 + y^2}{(z + i z_c )^2}} $$



$$ \mathbf{\hat{R}} = \frac{\mathbf{R}} {|\mathbf{R}|} = \frac{x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}}{R} $$


$\mathbf{J}_N$ is orientated along the x-axis with magnitude $J_x$.  Thus

$$ \mathbf{J}_N = J_x \mathbf{\hat{x}} $$

$$ \mathbf{J}_N \cdot\mathbf{\hat{R}} = \frac{J_x x}{R} $$


$$ (\mathbf{J}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}} = \frac{J_x x}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)

and

$$ \mathbf{J}_N \times \mathbf{\hat{R}} = J_x\mathbf{\hat{x}} \times \frac{x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}}{R} $$

$$ \mathbf{J}_N \times \mathbf{\hat{R}} = - J_x\Bigg(\frac{z + i z_c}{R} \mathbf{\hat{y}} - \frac{y}{R} \mathbf{\hat{z}} \Bigg) = -\frac{J_x (z + i z_c)}{R} \mathbf{\hat{y}} + \frac{J_x y}{R} \mathbf{\hat{z}}$$

Thus the fields due to an electric current element at a complex source point, aligned with the x-axis are:

$$ \mathbf{E}_N = \frac{k^2}{4\pi} J_x \mathrm{\hat{x}} \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + \frac{J_x x}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg) $$

$$ \mathbf{H}_N = \frac{k^2}{4\pi} J_x\Bigg(\frac{y}{R} \mathbf{\hat{z}} - \frac{z + i z_c}{R} \mathbf{\hat{y}} \Bigg)\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg) e^{-ikR} $$

Similarly to the calculations for $\mathbf{J}$ above, for a magnetic current element $\mathbf{M}$ orientated along the y-axis:

$$ (\mathbf{M}_N\cdot\mathbf{\hat{R}})\mathbf{\hat{R}} = \frac{M_y y}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)

and

$$ \mathbf{M}_N \times \mathbf{\hat{R}} = M_y \Bigg(  \frac{z + i z_c}{R} \mathbf{\hat{x}} - \frac{x}{R} \mathbf{\hat{z}}\Bigg) $$

Thus the fields due to an Magnetic current element at a complex source point, aligned with the x-axis are:

$$ \mathbf{E}_N = - \frac{k^2}{4\pi} M_y\Bigg(\frac{z + i z_c}{R} \mathbf{\hat{x}} - \frac{x}{R} \mathbf{\hat{z}} \Bigg)\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg) e^{-ikR} $$

$$ \mathbf{H}_N = \frac{k^2}{4\pi} \Bigg[M_y \mathrm{\hat{y}} \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + \frac{M_y y}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

# Symmetrization

Our normalized fields should obey the duality principle, where the fields are invariant under the substitution

$$ \mathbf{E}_N \rightarrow -\mathbf{H}_N $$

$$ \mathbf{H}_N \rightarrow  \mathbf{E}_N $$

$$ \mathbf{J}_N \rightarrow - \mathbf{M}_N $$

$$ \mathbf{M}_N \rightarrow \mathbf{J}_N $$

We make the substitutions in the first set of equations:

$$ \mathbf{H}_N = - \frac{k^2}{4\pi} \Bigg[M_y \mathrm{\hat{y}} \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) - \frac{M_y y}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)\Bigg] $$

$$ \mathbf{E}_N = - \frac{k^2}{4\pi} M_y \Bigg(\frac{z + i z_c}{R} \mathbf{\hat{x}} - \frac{x}{R} \mathbf{\hat{z}} \Bigg)\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg) e^{-ikR} $$

And add them to the first set of equations

$$ \mathbf{E}_N = \frac{k^2}{4\pi} \Bigg[J_x \mathrm{\hat{x}} \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + \frac{J_x x}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg) $$

$$ - M_y \Bigg(\frac{z + i z_c}{R} \mathbf{\hat{x}} - \frac{x}{R} \mathbf{\hat{z}} \Bigg)\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg] e^{-ikR} $$

$$ \mathbf{H}_N = \frac{k^2}{4\pi} \Bigg[M_y \mathrm{\hat{y}} \Bigg( \frac{-i}{kR} - \frac{1}{(kR)^2} + \frac{i}{(kR)^3}\Bigg) + \frac{M_y}{Z} \frac{y}{R^2} \big(x \mathbf{\hat{x}} + y \mathbf{\hat{y}} + (z + i z_c) \mathbf{\hat{z}}\big)\Bigg( \frac{i}{kR} + \frac{3}{(kR)^2} - \frac{3i}{(kR)^3} \Bigg)$$

$$ + J_x\Bigg(\frac{z + i z_c}{R} \mathbf{\hat{y}} - \frac{y}{R} \mathbf{\hat{z}} \Bigg)\Bigg(\frac{i}{kR}+\frac{1}{(kR)^2}\Bigg)\Bigg]  e^{-ikR} $$

These fields will be invariant if $M_y = -J_x$.

# Symmetrized Gaussian beam in GRASP normalization

Our forward time direction fields for a symmetrized Gaussian beam are therefore:

$$ E_x = \frac{J_x k^2}{4 \pi} e^{-ikr}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = - \frac{J_x k^2}{4 \pi} e^{-ikr}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ E_z = - \frac{J_x k^2}{4 \pi} e^{-ikr}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = -\frac{J_x k^2}{4 \pi } e^{-ikr}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{J_x k^2}{4 \pi } e^{-ikr}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = - \frac{J_x k^2}{4 \pi } e^{-ikr}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{y}{r}\Bigg] $$

# Normalization

The above equations give results with very large or small magnitudes for the fields due to the imaginary part of $r$ giving the exponential of a real number in $e^{±i k r}$. We also would like the phase of the E-field to be zero at the origin.  Per Cullen & Yu, we therefore multiply the fields by a constant factor $ - e^{\mp k z_c} $, so that:

$$ e^{-i k r} \rightarrow e^{- i k r - k z_c} $$

We wish to normalize the power in the beam to $4\pi \sqrt{\mathrm{Watt}}$.  To do so, we must multiply the beam by a normalizing factor



$$ \frac{- 4 \pi z_c}{2 k w_0} e^{- kz_c} =  $$


$$ z_c = \pi w_0^2 / \lambda = k w_0^2 / 2 $$

$$ \frac{4 \pi w_0}{\sqrt{2}} e^{- k^2 w_0^2 / 2}$$

# Final form

The final form for the fields of a Gaussian beam, normalized to $4 \pi \sqrt{\mathrm{W}}$ are:

$$ E_x = \frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ E_y = \frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ E_z = \frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{x}{r}\Bigg] $$

$$ H_x = -\frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2}\bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{xy}{r^2} $$

$$ H_y = \frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2} e^{ikr}\Bigg[\bigg(\frac{-i}{(kr)} - \frac{1}{(kr)^2} + \frac{i}{(kr)^3}\bigg) + \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{y^2}{r^2} - \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{z + i z_c}{r}\Bigg] $$

$$ H_z = - \frac{w_0 k^2}{\sqrt{2}} e^{-ikr - k^2w_0^2/2} e^{ikr}\Bigg[ \bigg(\frac{i}{(k r)} + \frac{3}{(kr)^2} - \frac{3i}{(kr)^3}\bigg)\frac{x(z + i z_c)}{r^2} + \bigg(\frac{i}{(kr)} + \frac{1}{(kr)^2}\bigg)\frac{y}{r}\Bigg] $$

```python
import numpy as np
import matplotlib.pyplot as plt

import scipy as sp
    
from PyPO.System import System
from PyPO.Enums import Units
```

## Define the Gaussian Beam

```python
# Setting up parameters for defining simulation.
# Source parameters and distances
lam = 10*Units.MM                # Wavelength of light in mm
k = 2*np.pi/lam
print(f"Wavenumber                : {k:7.3f} mm")

w0 = 5*lam  # Gaussian beamwaist

r_source = 2.5*w0               # Radius of disk in mm
print(f"Radius of source grid     : {r_source:7.3f} mm")

zc = np.pi*w0**2 / lam # Confocal distance in mm
print(f"Confocal distance         : {zc:7.3g} mm")

E0 = np.sqrt(8/(w0**2)) # Normalise to 4pi Watts
print(f"Field normalization       : {E0:7.3g} √Watt")

theta_0 = np.rad2deg(np.atan(lam/(np.pi*w0))) # Calculate far-field divergence angle
print(f"Farfield divergence angle : {theta_0:7.3f}°")

fwhm = 1.18*theta_0
print(f"Farfield FWHM angle       : {fwhm:7.3f}°")

# Calculate the size at the test point
z_grid = 100*lam
print(f"Near-field grid distance     : {z_grid:7.3f} mm")

w_grid = w0*np.sqrt(1 + ((lam*z_grid)/(np.pi*w0**2))**2)
print(f"Beamwidth at near-field grid : {w_grid:7.3f} mm")
```

```python
# Create a system

s = System()
```

```python
# Create a source grid
source_plane = {
        "name"      : "source_plane",
        "gmode"     : "uv",
        # "lims_x"    : np.array([-r_disk, r_disk]),
        # "lims_y"    : np.array([-r_disk, r_disk]),
        "lims_u"    : np.array([0, r_source]),
        "lims_v"    : np.array([0, 360]),
        "gridsize"  : np.array([101, 360])
        }

s.addPlane(source_plane)

s.translateGrids("source_plane", np.array((0, 0, z_grid)))

source_grid = s.generateGrids("source_plane")
```

# Source currents

We now calculate the source current distribution on a (real) plane that will give rise to these fields.

We want the radiated fields at the plane to be equal to the incident fields.  The induced currents on a surface that give rise to a radiated fields that is equal to the incident fields are those of a perfect electrical conductor for the electric current elements, and those of a perfect magnetic conductor for the magnetic current elements.  In each case, the respective equivalent currents are:

$$ \mathbf{J} = \mathbf{n} \times \mathbf{H} $$

$$ \mathbf{M} = -\mathbf{n} \times \mathbf{E} $$


For a plane surface normal to the z-axis, this becomes:

$$ \mathbf{J} = -H_y \mathbf{\hat{x}} + H_x \mathbf{\hat{y}} $$

$$ \mathbf{M} = E_y \mathbf{\hat{x}} - E_x \mathbf{\hat{y}} $$
