```python
import numpy as np

from PyPO.Checks import check_vecGPODict, calc_beam
from PyPO.Enums import Units
```

```python
# Set up a Gaussian beam with known parameters

lam = 1.0 * Units.MM

w0 = 5.0*lam
z = 10.0*lam

w_ref = w0*np.sqrt(1 + ((lam*z)/(np.pi*w0**2))**2)
R_ref = z*(1 + ((np.pi*w0**2)/(lam*z))**2)
```

```python
print(f"Input lambda : {lam:.6f} mm")
print(f"Input w0     : {w0:.6f} mm")
print(f"Input z      : {z:.6f} mm")
print(f"Reference w  : {w_ref:.6f} mm")
print(f"Reference R  : {R_ref:.6f} mm")
```
